***Broadcasting***

```python
l = [10,20,30,40]
arr = np.array(l)
```

```python
for i in l :
    print(i+10)
```

```python
arr +10
```

```python
arr * 2
```

***Matrix Operations***

```python
mat1 = np.arange(1,10).reshape((3,3))
mat1
```

```python
mat2 = np.arange(11,20).reshape(3,3)
mat2
```

```python
mat1*mat2
```

```python
np.dot(mat1,mat2)
```

```python
cat1 = np.array([[1,2],
        [3,4]])
```

```python
cat2 = np.array([[5,6],
        [7,8]])
```

```python
cat1*cat2
```

```python
cat1 @ cat2
```

```python
np.dot(cat1,cat2)
```
