__NUMPY INDEXING AND SLICING OF VECTORS__

```python
import numpy as np
```

```python
arr = np.arange(11,21)
arr
```

```python
arr[9]
```

```python
arr[:2]
```

```python
arr[:]
```

```python
arr_slice = arr[1:8:2]
arr_slice
```

__NUMPY INDEXING AND SLICING OF MATRIX__

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[0][1]
```

```python
arr[0,0]
```

```python
arr[0,4]
```

```python
arr[5,4]
```

```python
arr[0:2,1:3]
```

```python
arr[3:,3:]
```

```python
arr[:,2]
```

__BOOLEAN INDEXING__

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr%2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
